#objective is to construct a dataframe containing information about individual contributions to your
candidate’s presidential committee. To that end you will need to perform the following activities.
1. Locate your candidate’s unique FEC candidate id via the FEC Open Candidate API.
2. Locate your candidate’s official presidential committee id by using your candidate’s id to
filter the committee master data (provided as a delimited file).
3. Isolate all the individual contributions to your candidate’s official presidential committee
(and hence your candidate) by filtering the Contributions by Individuals data (provided as
two delimited files) based on your candidate’s presidential committee id.

```python
"""
__author__ = "Jash Enakulapati"
__credits__ = "Dr. Roberts"
__version__ = "1.0"
__maintainer__ = "Jash Enakulapati"
__email__ = "enakulapatij.edu"
__status__ = "Dev"
__Candidate_Analyzed__ = "Elizabeth Warren"
"""
import requests
import pandas as pd
from requests.exceptions import HTTPError

FEC_URL = "https://api.open.fec.gov/v1/candidates"
#API_KEY = "<<Your API Key and pass it in Params or "DEMO_KEY" in the params>>"

HEADERS = {
    'Content-Type': 'application/json',
    'Accept-Language': 'en_US',
}

PARAMS = {
    # Pass the parameters in the https request
    "party":"DEM",
    "name":"Elizabeth Warren",
    # For Testing - Used Other Candidates
    #"name":"Booker, Cory",
    #"name":"Tom Steyer",
    #"name":"Sanders, Bernie",
    "api_key":"DEMO_KEY",
    "office":"P",
    "election_years": "2020"
}

try:
    response = requests.get(FEC_URL, params=PARAMS, headers=HEADERS)
    response.raise_for_status()
except HTTPError as http_err:
    print(f'HTTP error occurred: {http_err}')
except Exception as err:
    print(f'Other error occurred: {err}')
else:
    print('GET Request Success!')
    response_data = response.json()
    candidate_details = response_data['results'][0]
    # Locate your candidate’s unique FEC candidate id via the FEC Open Candidate API.
    print(f"Elizabeth Warren's Candidate ID is: {candidate_details['candidate_id']}")
    

    
```

```python
CM_HEADER = pd.read_csv("cm_header_file.csv")
#print(CM_HEADER)
CM_DF = pd.read_csv("cm20.txt", sep="|", names = CM_HEADER)
# Locate your candidate’s official presidential committee id by using your candidate’s id to 
# filter the committee master data (provided as a delimited file).
committee_id = CM_DF[CM_DF.CAND_ID == candidate_details['candidate_id']].CMTE_ID.values
print(f"Elizabeth Warren's Committee ID is: {committee_id[0]}")
```

```python
# Isolate all the individual contributions to your candidate’s official presidential committee
# (and hence your candidate) by filtering the Contributions by Individuals data (provided as
# two delimited files) based on your candidate’s presidential committee id.
INDV_CONTR_HDR = pd.read_csv("indiv_header_file.csv", low_memory=False)
#print(INDV_CONTR_HDR)
def read_file(ipfilename, opfilename):
    with open(ipfilename) as f:
        with open(opfilename, "w") as s:
            for line in f:
                #Pull only those lines which starts with the my candidate's - Committee_Id
                if line.startswith(committee_id[0]):
                    s.write(line)
read_file("itcont_2020_20190629_20190908.txt", "filtered_itcont_2020_20190629_20190908.txt")
read_file("itcont_2020_20190909_20191120.txt", "filtered_itcont_2020_20190909_20191120.txt")
INDV_CONTRI_JUN_TO_SEP = pd.read_csv("filtered_itcont_2020_20190629_20190908.txt", sep="|", header=None)
INDV_CONTRI_SEP_TO_NOV = pd.read_csv("filtered_itcont_2020_20190909_20191120.txt", sep="|", header=None)
# print(INDV_CONTRI_JUN_TO_SEP.head(10))
# print(INDV_CONTRI_SEP_TO_NOV.tail(10))
#INDV_CONTRI = pd.concat([INDV_CONTRI_JUN_TO_SEP, INDV_CONTRI_SEP_TO_NOV],axis=0)
print(INDV_CONTRI_JUN_TO_SEP.shape)
print(INDV_CONTRI_SEP_TO_NOV.shape)
# Combine both quarter dataframes into a single dataframe
INDV_CONTRI = INDV_CONTRI_JUN_TO_SEP.append(INDV_CONTRI_SEP_TO_NOV, ignore_index = True)
print(INDV_CONTRI.shape)
print(INDV_CONTRI.head(5))
print(INDV_CONTRI.tail(5))
```

```python
INDV_CONTRI.columns.astype('str')
INDV_CONTRI.columns = INDV_CONTR_HDR.columns
# Slice only selected columns:
REQ_COLS = {'CMTE_ID':str, 'TRANSACTION_PGI':str, 'ENTITY_TP':str, 'NAME':str, 'CITY':str, 
            'STATE':str, 'ZIP_CODE':str, 'OCCUPATION':str, 'TRANSACTION_DT':str, 'TRANSACTION_AMT':float}
temp_df = INDV_CONTRI[REQ_COLS.keys()]
INDV_CONTRI_DF = temp_df.astype(REQ_COLS)
print(INDV_CONTRI_DF.dtypes)
```

Restricting Rows Using Filters
Filter the individual contributions data based on the following criteria:
1. Contributions made during the month of September year of 2019
2. Your candidate’s presidential committee id
3. Transaction amounts > 0
4. Contributions made by entities of type individual

```python
# Total Dataframe count before filtering
print(f"Total DataFrame size before applying the filter is: {INDV_CONTRI_DF.shape}\n")

# Apply filter to pull september individual contributors with transaction amount > 0.
# Note: Committee ID filter is not applied since this was taken care when sliced files were created in 3rd Cell.
sep_contributions = INDV_CONTRI_DF.loc[(INDV_CONTRI_DF['TRANSACTION_DT'].str.match(r'[0]*[9][0-9][0-9]2019')) &\
                                       (INDV_CONTRI_DF['TRANSACTION_AMT'] > 0) &\
                                       (INDV_CONTRI_DF['ENTITY_TP'] == 'IND')]

# 1. Contributions made during the month of September year of 2019
print(f"1. The number of contributions made during September 2019 to presidential candidate: {candidate_details['name']} is {sep_contributions.shape}\n")

# 2. Candidate’s presidential committee id
print(f"2. {candidate_details['name']}'s presidential committee id is: {committee_id[0]}\n")

# 3. Transaction amounts > 0
print(f"3. Number of transactions that are > 0 are: {sep_contributions[sep_contributions.TRANSACTION_AMT > 0].shape}\n")

# 4. Contributions made by entities of type individual
print(f"4. Number of contributions made by entities of type individuals are: {sep_contributions[sep_contributions.ENTITY_TP == 'IND'].shape}")
```

Other Transformations:
In the ethnicity and gender classification requirements that follow, you need access to contributors first
and last names.

```python
# Split name into first name and last name using the split function with seperator as comma
cols = {0:'last_name',1:'first_name'}
sep_contributions['NAME'] =  sep_contributions['NAME'].str.replace(r'\s\w\.',"")
FULL_NAME = sep_contributions['NAME'].str.split(',', expand=True).rename(columns=cols)
FULL_NAME['last_name'] = FULL_NAME['last_name'].str.strip(",")
FULL_NAME['first_name'] = FULL_NAME['first_name'].str.strip()
FULL_NAME['first_name'] = FULL_NAME['first_name'].str.strip(".")
FULL_NAME['first_name'] = FULL_NAME['first_name'].str.split(' ', expand=True)
# Ensure that there is no null values in firstname or lastname
print(FULL_NAME.isnull().any())
```

Ethnicity Classification
In this step you will be using the surnames from your FEC dataframe to impute the ethnicity of individual
contributors by importing isys613_project. The isys613_project module exposes two functions related to ethnicity prediction - predict_ethnicity_1 and predict_ethnicity_2. Each function uses a different classification methodology.

```python
# pip install -U ethnicolr==0.3.0
# import Dr. Roberts module for ethinicity classification using tensor flow and gender classification using nltk.
import isys613_project as tf
# Ethinicity Classification Method using Census data
ethinicity_est_1 = tf.predict_ethnicity_1(FULL_NAME, 'last_name')
# Ethinicity Classification Method using Florida registration data
ethinicity_est_2 = tf.predict_ethnicity_2(FULL_NAME, 'last_name')

```

```python
ethinicity_est_2.rename(columns={"asian": "asian_2", "black": "black_2", "hispanic": "hispanic_2", "white": "white_2"}, \
                        inplace=True)
# Merge the the estimation deduced by two classification methods
merge_ethinicity = pd.concat([ethinicity_est_1, ethinicity_est_2], axis = 1)
avg_ethinicity =  pd.DataFrame()
# Take the mean value for all 4 ethinicities
avg_ethinicity['asian'] = merge_ethinicity[['asian','asian_2']].mean(axis=1)
avg_ethinicity['black'] = merge_ethinicity[['black','black_2']].mean(axis=1)
avg_ethinicity['hispanic'] = merge_ethinicity[['hispanic','hispanic_2']].mean(axis=1)
avg_ethinicity['white'] = merge_ethinicity[['white','white_2']].mean(axis=1)
# Ensure the row count is same before and after concatination.
print(ethinicity_est_1.shape)
print(ethinicity_est_2.shape)
print(merge_ethinicity.shape)
print(avg_ethinicity.shape)
```

```python
# Match to one etinicity if estimation value is greater than 50% else Unknown
avg_ethinicity.loc[avg_ethinicity[['asian', 'black','hispanic','white']].max(axis=1).gt(0.50), 'ethinicity'] = avg_ethinicity[['asian', 'black','hispanic','white']].idxmax(axis=1)
avg_ethinicity.loc[avg_ethinicity[['asian', 'black','hispanic','white']].max(axis=1).le(0.50), 'ethinicity'] = "Unknown"
FULL_NAME['ethinicity_avg'] = avg_ethinicity['ethinicity']
full_df = avg_ethinicity.merge(FULL_NAME, left_index=True, right_index=True)
print(full_df.head(10))
```

Gender Classification
In this step you will be using the given name (first name) from your FEC dataframe to impute the gender
of individual contributors. Use predict_gender function as isys613_project

```python
# Classify the Gender by calling the predict_gender function
FULL_NAME['GENDER'] = tf.predict_gender(FULL_NAME, 'first_name')
print(FULL_NAME.head(10))
```

Analysis – Proof of Concept:

```python
import matplotlib.pyplot as plt
print(f"Analysis 1 - Total Contribution to Elizabeth Warren's Campaign in September is: {INDV_CONTRI_FINAL['TRANSACTION_AMT'].sum()}\n")
Contri_by_state = pd.DataFrame()
print("Analysis 1.1 - Top 5 Contributing states in September are as follows:\n")
Contri_by_state = INDV_CONTRI_FINAL.groupby(['STATE'])['TRANSACTION_AMT'].sum().reset_index(name='Total_Contribution') \
      .sort_values(by = ['Total_Contribution'], ascending=False).head(5)
print(Contri_by_state.to_string(index=False))
plt.figure(figsize=(10,8))
ax1 = plt.subplot(121, aspect='equal')
Contri_by_state.plot(kind='pie', y = 'Total_Contribution', ax=ax1, autopct='%1.1f%%', \
                     startangle=90, shadow=False, labels=Contri_by_state['STATE'], legend = False, fontsize=14)
plt.show()
print("Analysis 1.2 - Top 5 Contributing Cities in September are as follows:\n")
Contri_by_city = INDV_CONTRI_FINAL.groupby(['CITY'])['TRANSACTION_AMT'].sum().reset_index(name='Total_Contribution') \
      .sort_values(by = ['Total_Contribution'], ascending=False).head(5)
print(Contri_by_city.to_string(index=False))
plt.figure(figsize=(10,8))
ax1 = plt.subplot(121, aspect='equal')
Contri_by_city.plot(kind='pie', y = 'Total_Contribution', ax=ax1, autopct='%1.1f%%', \
                     startangle=90, shadow=False, labels=Contri_by_city['CITY'], legend = False, fontsize=14)
plt.show()
```

```python
print("Analysis 2 - Contributions by Gender in September are as follows:\n")
gender_contri = INDV_CONTRI_FINAL.groupby(['GENDER'])['TRANSACTION_AMT'].sum().reset_index(name='Total_Contribution')
print(gender_contri.to_string(index=False))
print("\nAnalysis 2.1 - Male & Female contributors in September are as follows:\n")
gender_count = INDV_CONTRI_FINAL.groupby(['GENDER'])['TRANSACTION_AMT'].count().reset_index(name='Count')
print(gender_count.to_string(index=False))
#print(INDV_CONTRI_FINAL['GENDER'].value_counts(normalize=True) * 100)
plt.figure(figsize=(10,8))
ax2 = plt.subplot(121, aspect='equal')
gender_count.plot(kind='pie', y = 'Count', ax=ax2, autopct='%1.1f%%', \
                     startangle=180, shadow=False, labels=gender_count['GENDER'], legend = False, fontsize=14)
plt.show()
```

```python
print("Analysis 3. Contributions by Top 10 Occupations in September are as follows:\n")
Ocup_Contri = INDV_CONTRI_FINAL.groupby(['OCCUPATION'])['TRANSACTION_AMT'].sum().reset_index()\
      .rename(columns = {'TRANSACTION_AMT': 'Total_Contribution'})\
      .sort_values(by = ['Total_Contribution'], ascending=False).head(10)
print(Ocup_Contri.to_string(index=False))
plt.figure(figsize=(16,12))
ax3 = plt.subplot(121, aspect='equal')
Ocup_Contri.plot(kind='pie', y = 'Total_Contribution', ax=ax3, autopct='%1.1f%%', \
                     startangle=160, shadow=False, labels=Ocup_Contri['OCCUPATION'], legend = False, fontsize=12)
plt.show()
```

```python
print("Analysis 4. Contribution by Ethinicity in September are as follows:\n")
Contri_by_Ethinicity = INDV_CONTRI_FINAL.groupby(['ethinicity_avg']).agg({'TRANSACTION_AMT':'sum','ENTITY_TP':'count'}) \
.reset_index().rename(columns={'ethinicity_avg':'Ethinicity','TRANSACTION_AMT':'Total_Contribution', 'ENTITY_TP':'Count'}) \
      .sort_values(by = ['Total_Contribution'], ascending=False)
print(Contri_by_Ethinicity.to_string(index=False))
plt.figure(figsize=(16,10))
ax3 = plt.subplot(121, aspect='equal')
Contri_by_Ethinicity.plot(kind='pie', y = 'Total_Contribution', ax=ax3, autopct='%.1f%%', \
                     startangle=140, shadow=False, labels=Contri_by_Ethinicity['Ethinicity'], legend = True, fontsize=12)
plt.show()
```
